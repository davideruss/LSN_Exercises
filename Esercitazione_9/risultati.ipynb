```python
import numpy as np
import matplotlib.pyplot as plt
```

## Risultati algoritmo genetico


```python
x, y= np.loadtxt("INPUT/circle.dat", unpack =True)
a = np.loadtxt("OUTPUT/bestc.dat", dtype= int)
x_f= x[a]
y_f = y[a]

cost = np.loadtxt("OUTPUT/l_avec.dat")
i = np.arange(len(cost))

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize= (19, 7))
ax1.scatter(x,y)
ax1.set_title("City positions")

ax2.plot(x_f, y_f)
ax2.set_title("Best path")

ax3.plot(i, cost)
ax3.grid(True)
ax3.set_title("Average length evolution")
ax3.set_xlabel("Number of epochs")
ax3.set_ylabel("Average lenght")

plt.tight_layout()
plt.show()
```

Da questi grafici possiamo osservare che l'algoritmo ha chiaramente trovato la miglior configurazione. Inoltre osserviamo che la lunghezza media degli individui della popolazione diminuisce all'aumentare delle epoche come ci aspettavamo, il rumore che osserviamo nel terzo grafico ci dice anche che il nostro algoritmo se non trova minimi locali "troppo profondi" dovrebbe riuscire ad uscirne e ad esplorare altre configurazioni.

```python
x, y= np.loadtxt("INPUT/square.dat", unpack =True)
a = np.loadtxt("OUTPUT/bests.dat", dtype= int)
x_f= x[a]
y_f = y[a]

cost = np.loadtxt("OUTPUT/l_aves.dat")
i = np.arange(len(cost))

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize= (16, 6))
ax1.scatter(x,y)
ax1.set_title("city positions")

ax2.plot(x_f, y_f)
ax2.set_title("Best path")

ax3.plot(i, cost)
ax3.grid(True)
plt.show()
```

Vediamo che con una popolazione di 100 individui e 5000 epoche riusciamo ad arrivare ad un percorso minimo di lunghezza 9.79002.
Anche in questo caso notiamo che la lunghezza media degli individui decresce correttamente all'amumentare delle epoche trascorse.
