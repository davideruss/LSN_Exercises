## Esercizio 1.1

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Grafico della media

```python

average, errors = np.loadtxt('Average.txt', unpack=True)

x= np.linspace(1, 10000, 100)
plt.errorbar(x, average, errors)
plt.xlabel('#throws')
plt.ylabel("$\\langle r \\rangle -0.5$")
plt.grid(True)
plt.title("Media del generatore di numeri casuali")
plt.show()
```

notiamo dal grafico che il valor medio della distribuzione è in accordo con 0.5 che è il valore che ci aspettiamo da una distribuzione uniforme in [0,1)

##  Grafico della deviazione standard

```python
average2, errors2 = np.loadtxt('devstd.txt', unpack=True)

x= np.linspace(1, 10000, 100)
plt.errorbar(x, average2, errors2)
plt.xlabel('#throws')
plt.ylabel("$\\sigma - 1/12$")
plt.grid(True)
plt.title("Deviazione std del generatore di numeri casuali")
plt.show()
```

Anche qua troviamo un risultato in accordo con il valore atteso.

# Grafico della distribuzione del $\chi ^2$

```python
data= np.loadtxt('chi^2.txt')
from scipy.stats import chi2

x = np.linspace(50,170)
plt.plot(x, chi2.pdf(x,100))
plt.hist(data, bins=40, range=(0, 150), density =True )
plt.title("Distribuzione del $\chi^2$")
plt.grid(True)
```

Il $\chi^2$ è correttamente distribuito attorno a 100.
