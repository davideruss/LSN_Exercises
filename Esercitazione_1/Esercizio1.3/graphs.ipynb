```python
import matplotlib.pyplot as plt
import numpy as np
```

# Esercizio 1.3
## Esperimento di Buffon

```python

average, errors = np.loadtxt('Buffon.txt', unpack=True)

x= np.linspace(1, 100000, 100)
plt.errorbar(x, average, errors)
plt.xlabel('#throws')
plt.ylabel("valore misurato di $\pi$")
plt.grid(True)
plt.axhline(y=np.pi, color='red', linestyle='--')
plt.title("Misura di $\pi$")
plt.show()

print("valore misurato = ", average[-1], "+/-" , errors[-1] )
```

il valore misurato è in accordo con il valore aspettato del pi greco
