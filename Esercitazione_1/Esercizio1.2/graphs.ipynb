```python
import matplotlib.pyplot as plt
import numpy as np
```

## Distribuzioni di Cauchy Lorentz

```python
L = [1, 2, 10, 100]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i in range(4):

    data = np.loadtxt(f'histo_Lorentz_{L[i]}.txt')

    axs[i].hist(data, bins=50, edgecolor='blue',range=(-5, 5), density=True)
    axs[i].set_title(f'Misura con media di {L[i]} valori')
    axs[i].set_xlabel('misura')
    axs[i].set_ylabel('Frequenza')
    axs[i].grid(True, alpha=0.3)
    axs[i].set_xlim(-5, 5)
        

plt.tight_layout()
plt.show()
```

Questa distribuzione rimane invariata e non tende ad una Gaussiana, quindi in questo caso non vale il teorema del limite centrale.

## Distribuzione Esponenziale

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i in range(4):

    data = np.loadtxt(f'histo_exp_{L[i]}.txt')
           
    axs[i].hist(data, bins=50, edgecolor='blue',range=(0, 2), density=True)
    axs[i].set_title(f'Misura con media di {L[i]} valori')
    axs[i].set_xlabel('misura')
    axs[i].set_ylabel('Frequenza')
    axs[i].grid(True, alpha=0.3)
    axs[i].set_xlim(0, 2)
    if(i ==3):
        dev_st = np.std(data)
        from scipy.stats import norm
        x = np.linspace(0,2, 100)
        y =norm.pdf(x, 1, dev_st)
        axs[i].plot(x, y)


plt.tight_layout()
plt.show()
```

In questo caso vediamo che vale il teorema del limite centrale e alla fine la distribuzione tende ad una gaussiana

## Distribuzione Uniforme

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i in range(4):

    data = np.loadtxt(f'histo_Uniform_{L[i]}.txt')
           
    axs[i].hist(data, bins=50, edgecolor='blue',range=(0, 1), density=True)
    axs[i].set_title(f'Misura con media di {L[i]} valori')
    axs[i].set_xlabel('misura')
    axs[i].set_ylabel('Frequenza')
    axs[i].grid(True, alpha=0.3)
    axs[i].set_xlim(0, 1)
    if(i ==3):
        dev_st = np.std(data)
        from scipy.stats import norm
        x = np.linspace(0,1, 150)
        y =norm.pdf(x, 0.5, dev_st)
        axs[i].plot(x, y)


plt.tight_layout()
plt.show()
```

Anche in questo caso il teorema del limite centrale vale, e mediando su tanti valori vediamo che la distribuzione tende ad una gaussiana.
