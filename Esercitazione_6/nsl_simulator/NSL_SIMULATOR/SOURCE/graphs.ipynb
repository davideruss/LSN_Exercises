# Esercitazione 6

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
```

# Energia interna, suscettibilità, calore specifico

```python
x=np.linspace(0.5, 2.0, 16)
cv = []
cv_err = []
susc=[]
susc_err=[]
energy=[]
energy_err=[]

for i in range(16):
    l, s = np.loadtxt(f"../OUTPUT/sim_type_3_h_0/specific_heat{15-i}.dat", unpack = True, usecols=(2, 3), skiprows=20)
    t, u = np.loadtxt(f"../OUTPUT/sim_type_3_h_0/susceptibility{15-i}.dat", unpack = True, usecols=(2, 3), skiprows=20)
    v, z = np.loadtxt(f"../OUTPUT/sim_type_3_h_0/total_energy{15-i}.dat", unpack = True, usecols=(2, 3), skiprows=20)
    cv.append(l)
    cv_err.append(s)
    susc.append(t)
    susc_err.append(u)
    energy.append(v)
    energy_err.append(z)

```

```python
fig, ax = plt.subplots(1, 3, figsize= (14, 8))
fig.suptitle("Gibbs method", fontsize =20)

ax[0].errorbar(x, cv, cv_err, fmt ='o', label="measure", markersize = 2)
ax[0].set_xlabel('Temperature')
ax[0].set_ylabel('specific heat')
ax[0].set_title("Specific Heat")
ax[0].plot(T, heat, label = "theory")
ax[0].grid(True)
ax[0].legend()

ax[1].errorbar(x, susc, susc_err, fmt ='o', label="measure", markersize = 2)
ax[1].set_xlabel('Temperature')
ax[1].set_ylabel('susceptibility')
ax[1].set_title("Susceptibility")
ax[1].plot(T,X , label = "theory")
ax[1].grid(True)
ax[1].legend()

ax[2].errorbar(x, energy, energy_err, fmt ='o', label="measure", markersize = 2)
ax[2].set_xlabel('Temperature')
ax[2].set_ylabel('Total Energy')
ax[2].set_title("Total Energy")
ax[2].plot(T,e , label = "theory")
ax[2].grid(True)
ax[2].legend()

plt.tight_layout()
plt.show()

```
