# Esercizio 3

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

average1, errors1 = np.loadtxt('Call-option_dir.dat', unpack=True)

x= np.linspace(1, 100, 100)
plt.errorbar(x, average1, errors1, label="Call-option direct")
plt.xlabel('number of blocks')
plt.ylabel("call-option price")
plt.grid(True)

average2, errors2 = np.loadtxt('Call-option_dis.dat', unpack=True)
plt.errorbar(x, average2, errors2, alpha= 0.7, label="Call-option discretized")
plt.axhline(y= 14.975, color= 'red', linestyle= '--', label="Expected value ")
plt.legend()
plt.title("Call-option")
plt.grid(True)
plt.show()

```

```python

average1, errors1 = np.loadtxt('Put-option_dir.dat', unpack=True)

x= np.linspace(1, 100, 100)
plt.errorbar(x, average1, errors1, label="Put-option direct")
plt.xlabel('Number of blocks')
plt.ylabel("Put-option price")
plt.grid(True)

average2, errors2 = np.loadtxt('Put-option_dis.dat', unpack=True)
plt.errorbar(x, average2, errors2, label= "Put-option discretized", alpha=0.7)
plt.axhline(y= 5.4595, color= 'red', linestyle= '--', label = "Expected value")
plt.title("Put-option")
plt.legend()
plt.grid(True)
plt.show()
```

Notiamo che in entrambi i casi i valori ottenuti sono compatibili con i valori aspettati con un errore in entrambi i metodi $\approx 1\%$.
