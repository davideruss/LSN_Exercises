```python
import matplotlib.pyplot as plt
import numpy as np
```

Grafico della media

```python

passi, r_rms, errori = np.loadtxt("RW_discreto.txt", unpack=True)

fig, ax = plt.subplots(figsize=(10, 6))

teoria = np.sqrt(passi)
ax.plot(passi, teoria, label='Teoria: $\sqrt{N}$', alpha=0.6)
ax.errorbar(passi, r_rms, yerr=errori, label='Simulazione')
ax.set_title("Random Walk discreto: distanza quadratica media in funzione del numero di passi")
ax.set_xlabel("Numero di passi (N)")
ax.set_ylabel("$\sqrt{\\langle |r|^2 \\rangle}$")
ax.grid()
ax.legend()

ax_zoom = ax.inset_axes([0.55, 0.15, 0.4, 0.4])
ax_zoom.errorbar(passi, r_rms, yerr=errori, fmt='o', markersize=2,capsize=3, alpha = 0.7)
ax_zoom.plot(passi, teoria, color='red', linewidth=2.5, zorder=5, alpha= 0.7)
zoom_start = len(passi)-20
zoom_end = len(passi)-10
ax_zoom.set_xlim(zoom_start, zoom_end)
y_min = np.min(r_rms[zoom_start:zoom_end] - errori[zoom_start:zoom_end])
y_max = np.max(r_rms[zoom_start:zoom_end] + errori[zoom_start:zoom_end])
ax_zoom.set_ylim(y_min - 0.5, y_max + 0.5)
ax_zoom.grid()
ax.indicate_inset_zoom(ax_zoom, edgecolor="black")

plt.tight_layout()
plt.show()


```
