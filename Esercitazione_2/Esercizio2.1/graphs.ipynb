```python
import matplotlib.pyplot as plt
import numpy as np
```

# Calcolo dell'integrale con MC

```python

average1, errors1 = np.loadtxt('integrali1.txt', unpack=True)

x= np.linspace(100, 10000, 100)
plt.errorbar(x, average1, errors1, alpha=0.7, label="without importance sampling")
plt.xlabel('#throws')
plt.ylabel("Valore integrale")
plt.grid(True)

average2, errors2 = np.loadtxt('integrali2.txt', unpack=True)
plt.errorbar(x, average2, errors2, label = "with importance sampling" )
plt.grid(True)
plt.title("Calcolo dell'integrale con MC")
plt.legend()
plt.show()
```

Possiamo vedere che il risultato ottenuto usando l'importance sampling risulta molto più preciso
